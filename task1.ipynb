```python
import matplotlib.pyplot as plt

# Age groups and their corresponding percentages
age_groups = ['0-14 years', '15-64 years', '65 years and over']
percentages = [11.4, 59.5, 29.1]

# Create the bar chart
plt.figure(figsize=(8, 5))
plt.bar(age_groups, percentages, color=['skyblue', 'lightgreen', 'salmon'])

# Add title and labels
plt.title("Japan's Population Distribution by Age Group (2023)")
plt.xlabel("Age Groups")
plt.ylabel("Percentage of Total Population")

# Display the chart
plt.show()
```
